```python
import seaborn as sns
```

```python
df_combined = pd.concat([
    pd.DataFrame({'price': df_ind['price'], 'airline': 'Indigo'}),
    pd.DataFrame({'price': df_vist['price'], 'airline': 'Vistara'}),
    pd.DataFrame({'price': df_air['price'], 'airline': 'AirIndia'}),
    pd.DataFrame({'price': df_go['price'], 'airline': 'GoFirst'}),
    pd.DataFrame({'price': df_aira['price'], 'airline': 'AirAsia'}),
    pd.DataFrame({'price': df_spice['price'], 'airline': 'SpiceJet'})
])
```

```python
plt.figure(figsize=(8,8))
sns.stripplot(x='airline', y='price', data=df_combined)
plt.title('Flight Price Distribution by Airline')
plt.figure(figsize=(12, 6))  # Set size before plotting
sns.stripplot(y="price", x="state", data=df)
plt.show()
plt.tight_layout()
plt.show()
```

 Does ticket price change based on the departure time and arrival time?

```python
df.groupby('departure_time')['price'].mean()
```

```python
sns.catplot(x='arrival_time',y='price', kind= 'bar', data = df)
plt.show()
```

```python
df.groupby('arrival_time')['price'].mean()
```

```python
sns.relplot( x = 'arrival_time', y = 'price', data = df, col = 'departure_time', kind = 'line')

plt.show()
```
